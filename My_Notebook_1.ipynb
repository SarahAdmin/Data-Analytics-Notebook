```python
import pandas as pd
housingdata = pd.read_csv('UK-HPI-full-file-2025-02.csv',encoding='latin-1')
columns_to_remove = ['Date','RegionName', 'AreaCode','IndexSA','CashPrice',
       'CashIndex', 'Cash1m%Change', 'Cash12m%Change', 'CashSalesVolume',
       'MortgagePrice', 'MortgageIndex', 'Mortgage1m%Change',
       'Mortgage12m%Change', 'MortgageSalesVolume', 'FTBPrice', 'FTBIndex',
       'FTB1m%Change', 'FTB12m%Change', 'FOOPrice', 'FOOIndex', 'FOO1m%Change',
       'FOO12m%Change',]
new_housingdata = housingdata.drop(columns_to_remove,axis=1)
final_housing_data = new_housingdata.corr()
print(final_housing_data.to_string())
```

```python
import pandas as pd
studentData = pd.read_json('student1.json')
columnstoDelete = ["'id_number'","'emailaddress'"]
new_data = studentData.drop(columnstoDelete,axis=1)
studentDF = new_data[(studentData["'coursecode'"] == 'CSC') | (studentData["'coursecode'"] == 'FST')|(studentData["'coursecode'"] == 'MST') ].dropna()
print(studentDF.to_string())
```

```python
import pandas as pd
studentData = pd.read_json('student1.json')
courseData = pd.read_json('course1.json')
studentsDF = studentData.merge(courseData)
columnstoDelete1 = ["'id_number'", "'emailaddress'","'duration'"]
new_students = studentsDF.drop(columnstoDelete1,axis=1)
business_students = new_students[(new_students["'coursecode'"] == 'BNS') | (new_students["'coursecode'"] == 'FIC')].dropna()
print(business_students.to_string())
```
